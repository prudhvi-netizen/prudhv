```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
from sklearn.linear_model import LogisticRegression
data1=pd.read_csv("set1.csv")
data2=pd.read_csv("set2.csv")
data1.head()
```

```python
data2.head()
```

```python
X_test=(data1.iloc[:,1:])
Y_test=(data1.iloc[:,:1])
X_train=(data2.iloc[:,1:])
Y_train=(data2.iloc[:,:1])
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
df1=pd.DataFrame(X_train)
df1= df1.convert_objects(convert_numeric=True)
df1=df1.fillna(df1.mean())
```

```python
df2=pd.DataFrame(X_test)
df2= df2.convert_objects(convert_numeric=True)
df2=df2.fillna(df2.mean())
```

```python
df1=scale(df1)
df2=scale(df2)
```

```python
Y_test=Y_test.apply(LabelEncoder().fit_transform)
```

```python
print(Y_test)
```

```python
Y_train=Y_train.apply(LabelEncoder().fit_transform)
```

```python
Y_train
```

```python
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(df1, Y_train.values.ravel())

#Predict the response for test dataset
y_pred = clf.predict(df2)
y_pred
```

```python
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```
